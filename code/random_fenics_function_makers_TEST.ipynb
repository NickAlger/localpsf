```python
import numpy as np
from random_fenics_function_makers import *
import matplotlib.pyplot as plt
from mesh_makers import circle_mesh
```

```python
mesh = circle_mesh(np.array([0.5,0.5]), 0.5, 1./1e-1)
V = fenics.FunctionSpace(mesh, 'CG', 2)
fenics.plot(mesh)
```

```python
u = randn_fenics_function(V)
fenics.plot(u)
```

```python
sfm = FenicsSmoothFunctionMaker(V)

u = sfm.random_smooth_function()
c0 = fenics.plot(u)
plt.colorbar(c0)
plt.title('u')

plt.figure()
c1 = fenics.plot(fenics.exp(0.1*u))
plt.colorbar(c1)
plt.title('exp(0.1*u)')

plt.figure()
c2 = fenics.plot(fenics.exp(u))
plt.colorbar(c2)
plt.title('exp(u)')

plt.figure()
c3 = fenics.plot(fenics.exp(10*u))
plt.colorbar(c3)
plt.title('exp(10*u)')
```
